### 50. Write a Python program to convert a string to datetime.
Sample String : Jul 1 2014 2:43PM 

Expected Output : 2014-07-01 14:43:00

```python
sample_dt = input('Enter datetime in string:')
```

```python
def str_to_dt(x):
    inter_dt = dt.datetime.strptime(x,'%b %d %Y %I:%M%p')
    output_dt = pd.to_datetime(inter_dt.strftime('%Y-%m-%d %H:%M:%S'))
    return output_dt
```

```python
str_to_dt(sample_dt)
```

```python
type(str_to_dt(sample_dt))
```

### 51. Write a Python program to subtract five days from current date.

Current Date : 2015-06-22

5 days before Current Date : 2015-06-17

```python
def sub_fivedays(x):
    inter_date = pd.to_datetime(x) + pd.DateOffset(days=-5)
    new_date = inter_date.strftime('%Y-%m-%d')
    return new_date
```

```python
sub_fivedays('2015-06-22')
```

### 54. Write a Python program to get days between two dates. 

```python
def daysbetweendates(x,y):
    day1 = pd.to_datetime(x)
    day2 = pd.to_datetime(y)
    diff = abs(day2-day1)
    return diff
```

```python
daysbetweendates('2011-09-20','2010-09-15')
```
