```python
3. Cross-Selling (Which products are selling together)
Hint: We need to find which of the top 10 combinations of products are selling together in
each transaction. (combination of 2 or 3 buying together)
```

```python
cross_selling = order_items[['order_id','product_id']].groupby('order_id')\
.agg({'product_id': lambda x: x.ravel().tolist()}).reset_index()\
.rename(columns = {'product_id':'items_bought_together'})
```

```python
cross_selling.head()
```

```python
cross_selling = cross_selling[cross_selling['items_bought_together'].apply(len) >1]\
.reset_index(drop=True)
```

```python
cross_selling.head()
```

```python
import itertools
```

```python
cross_selling.items_bought_together[[0]]
```

```python
list(itertools.combinations(cross_selling.items_bought_together[2],2))
```

```python
combinations_list = []
for row in cross_selling.items_bought_together:
    combinations = list(itertools.combinations(row,2))
    combinations_list.append(combinations)
```

```python
combination_counts = pd.Series(combinations_list).explode().reset_index(drop=True)
```

```python
combination_counts.value_counts().head(10)
```

```python
combinations_list = []
for row in cross_selling.items_bought_together:
    combinations = list(itertools.combinations(row,3))
    combinations_list.append(combinations)
```

```python
combination_counts = pd.Series(combinations_list).explode().reset_index(drop=True)
```

```python
combination_counts.value_counts().head(10)
```
